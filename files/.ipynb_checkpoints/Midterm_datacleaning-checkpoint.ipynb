'''As the electricity price is for Spain as a whole,the weather must be analysed for Spain as a whole as well. 
The weather is given for 5 cities. These major cities are located in different parts of Spain
Madrid - Center, Bilbao - North, Barcelona - North-West, Valencia -West, Seville - South-
Therefore, this analysis is based on the assumption that an average of weather data of these cities can roughly be 
considered to be representative for Spain as a whole.'''

```python
# Using the groupby function to find the averages (mode for 'weather_main') for all values for one single date.
weather_data = weather_dirt.groupby(['time']).agg({'temp':np.mean,'pressure':np.mean,'humidity':np.mean,'wind_speed':np.mean,'rain_1h':np.mean,'snow_3h':np.mean,'clouds_all':np.mean,'weather_main':st.mode}).reset_index()
```

```python
weather_data.head()
```

```python
# Converting temp from Kelvin to Celsius. This will be useful when visualising data for presentation.
def kelvin_to_celsius(temp):
    temp_celsius = temp - 273.15
    return temp_celsius
```

```python
weather_data['temp'] = weather_data['temp'].apply(kelvin_to_celsius)
```

## Hypothesis Testing

```python
data_ht = data.copy()
```

```python
# For hypothesis testing, numeriiiicals need to be converted into categories. Lets start with temp
ic2_labels = ['Low', 'Moderate', 'High', 'Very High']
data_ht['ic2_category_quartiles'] = pd.qcut(data_ht['IC2'],4,labels=ic2_labels) # quartiles as boundaries
data_ht[['IC2','ic2_category_quartiles']].head(20)
```

```python
temp_labels = ['Low', 'Moderate', 'High', 'Very High']
data_ht['temp_category'] = pd.cut(data_ht['temp'],4,labels=temp_labels)
#data_ht[['temp','temp_category']].head(20)
```

```python
data_crosstab = pd.crosstab(data_ht['temp_category'], data_ht['price_actual'], margins = False)
data_crosstab
```

```python
st.chi2_contingency(data_crosstab) # The p value is roughly 0,00001609 and therefore H0 has to be reject as there is a relation to electricity prices.
```

```python
# Now for wind_speed
wind_labels = ['Weak', 'Moderate', 'Strong', 'Very Strong']
data_ht['wind_category'] = pd.cut(data_ht['wind_speed'],4,labels=wind_labels)
```

```python
data_crosstab = pd.crosstab(data_ht['wind_category'], data_ht['price_actual'], margins = False)
data_crosstab
```

```python
st.chi2_contingency(data_crosstab) # There is good evidence that wind speed and electricity prices are related
```

```python
# Now for clouds_all
cloud_labels = ['very clear', 'clear', 'cloudy', 'very cloudy']
data_ht['cloud_category'] = pd.cut(data_ht['clouds_all'],4,labels=cloud_labels)
```

```python
data_crosstab = pd.crosstab(data_ht['cloud_category'], data_ht['price_actual'], margins = False)
data_crosstab
```

```python
st.chi2_contingency(data_crosstab)
```
